```python
import numpy as np
import matplotlib.pyplot as plt

phi = lambda z: np.exp(-z**2)
t = np.linspace(0,1,1000)
plt.figure(1)
plt.clf()

N= 50

for kk in range(N):
  plt.plot(t, phi(N*t - (kk + 1) + 0.5))
```

```python
import scipy.integrate as integrate

f = lambda z: (z < .25)*(4*z) + (z >= 0.25)*(z < 0.5)*(-4*z+2) - (z>= 0.5)*np.sin(14*np.pi*z)

int_list = []

for jj in range(N):
  f_phik = lambda z: f(z)*phi(N*z - jj + 0.5)

  int_list.append(integrate.quad(f_phik, 0, 1)[0])
```

```python
print(int_list)
int_list = np.array(int_list)
print(int_list)
```

```python
Gram_m = []

for ii in range(N):
  row = []
  for jj in range(N):
    f_phik = lambda z: phi(N*z - ii + 0.5)*phi(N*z - jj + 0.5)
    row.append(integrate.quad(f_phik, 0, 1)[0])

  Gram_m.append(row)

print(Gram_m)

Gram_m = np.array(Gram_m)

print(Gram_m)
```

```python
# Solve the system of equations
print(Gram_m.shape)
print(int_list.shape)

x = np.linalg.solve(Gram_m, int_list)

print('x',x)

# Print the solution
print("Solution:")
print(f"x1 = {x[0]}")
print(f"x2 = {x[1]}")
print(f"x3 = {x[2]}")
print(f"x4 = {x[3]}")
print(f"x5 = {x[4]}")
```

```python
#N = 5  # Change N to the desired number of entries
a = []

for i in range(N):
    x_value = x[i]
    a.append(x_value)
```

```python
y = np.zeros(1000)
phi = lambda z: np.exp(-z**2)

for jj in range(N):
  y = y+a[jj]*phi(N*t - (jj) + 0.5)

plt.figure(dpi=300)

plt.plot(t,y,label='approximated function')

# Define the function f(z)
f = lambda z: (z < .25)*(4*z) + (z >= 0.25)*(z < 0.5)*(-4*z+2) - (z>= 0.5)*np.sin(14*np.pi*z)
z_values = np.linspace(0, 1, 1000)
f_values = f(z_values)

# Plot the function
plt.plot(z_values, f_values, label='true function')
plt.xlabel('z')
plt.ylabel('f(z)')
plt.title('Plot with N=50')
plt.legend()
plt.grid(True)
plt.show()
```
